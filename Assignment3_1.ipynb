```python
import pandas as pd
import networkx as nx
import matplotlib
#matplotlib.use('Agg')
import matplotlib.pyplot as plt
# %matplotlib inline
```

# Facebook Network

```python
df = pd.read_csv('facebook.txt', sep=" ")
```

**Creating The Graph**

```python
G = nx.from_pandas_edgelist(df,'src','des',edge_attr=None,create_using=nx.DiGraph())
nx.info(G)
```

**Degree Centrality**

```python
degreeCentrality = nx.degree_centrality(G)
```

**Plot Degree Centrality for all nodes**

```python
lists = sorted(degreeCentrality.items())
x, y = zip(*lists)
plt.plot(x,y)
plt.ylabel('Degree Centrality')
plt.xlabel('Node ID')
plt.title('Degree Centrality of Facebook Dataset')
plt.show() 
```

**Closeness Centrality**

```python
closenessCentrality = nx.closeness_centrality(G)
```

**Plot Closeness Centrality for all nodes**

```python
lists = sorted(closenessCentrality.items())
x, y = zip(*lists)
plt.plot(x,y)
plt.ylabel('Closeness Centrality')
plt.xlabel('Node ID')
plt.title('Closeness Centrality of Facebook Dataset')
plt.show() 
```

**Betweenness Centrality**

```python
betweennessCentrality = nx.betweenness_centrality(G)
```

**Plot Betweenness Centrality for all nodes**

```python
lists = sorted(betweennessCentrality.items())
x, y = zip(*lists)
plt.plot(x,y)
plt.ylabel('Betweenness Centrality')
plt.xlabel('Node ID')
plt.title('Betweenness Centrality of Facebook Dataset')
plt.show() 
```

**Left Eigenvector Centrality**

```python
leftEigenvectorCentrality = nx.eigenvector_centrality_numpy(G, max_iter=100000000)
```

**Plot Left Eigenvector Centrality for all nodes**

```python
lists = sorted(leftEigenvectorCentrality.items())
x, y = zip(*lists)
plt.plot(x,y)
plt.ylabel('Left Eigenvector Centrality')
plt.xlabel('Node ID')
plt.title('Left Eigenvector Centrality of Facebook Dataset')
plt.show() 
```

**Right Eigenvector Centrality**

```python
rightEigenvectorCentrality = nx.eigenvector_centrality_numpy(G.reverse(), max_iter=100000000)
```

**Plot Right Eigenvector Centrality for all nodes**

```python
lists = sorted(rightEigenvectorCentrality.items())
x, y = zip(*lists)
plt.plot(x,y)
plt.ylabel('Right Eigenvector Centrality')
plt.xlabel('Node ID')
plt.title('Right Eigenvector Centrality of Facebook Dataset')
plt.show() 
```
