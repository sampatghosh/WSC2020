```python
import pandas as pd
import networkx as nx
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

# Facebook Network

```python
df = pd.read_csv('facebook.txt', sep=" ")
```

**Creating The Graph**

```python
G = nx.from_pandas_edgelist(df,'src','des',edge_attr=None,create_using=nx.DiGraph())
nx.info(G)
```

**Degree Centrality**

```python
degreeCentrality = nx.degree_centrality(G)
```

**Plot Degree Centrality Distribution**

```python
d = dict()
for i in degreeCentrality.values():
    if(d.get(i,0)==0):
        d[i] = 1;
    else:
        d[i] = 1 + d[i]
lists = sorted(d.items())
cen, freq = zip(*lists)  
plt.figure(figsize=(10,5))
plt.plot(cen,freq)
plt.ylabel('Frequency')
plt.xlabel('Degree Centrality')
plt.title(' Distribution of Degree Centrality in Facebook Network(Normal scale)')
plt.show()
plt.figure(figsize=(10,5))
plt.plot(cen,freq)
plt.xscale('log')
plt.yscale('log')
plt.ylabel('Frequency')
plt.xlabel('Degree Centrality')
plt.title(' Distribution of Degree Centrality in Facebook Network(Logarithmic scale)')
plt.show()
```

**Closeness Centrality**

```python
closenessCentrality = nx.closeness_centrality(G)
```

**Plot Closeness Centrality Distribution**
