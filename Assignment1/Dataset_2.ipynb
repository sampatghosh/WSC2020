```python
import pandas as pd
import networkx as nx
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/My Drive/M.Tech NITK/Sem 2/WSC/Assignments/facebook.txt', sep=" ") 
```

```python
G = nx.from_pandas_edgelist(df,'src','des',edge_attr=None)
```

```python
#degree-distribution
degs = {}
for n in G.nodes ():
    deg = G.degree (n)
    if deg not in degs :
        degs [deg ] = 0
    degs [deg] += 1
items = sorted ( degs . items ())
fig = plt.figure()
ax = fig.add_subplot (111)
ax.plot([k for (k,v) in items ], [v for (k,v) in items ])
#ax. set_xscale('log')
#ax. set_yscale('log')
plt.title("Facebook Degree Distribution")
plt.show()
```

```python
#diameter
nx.diameter(G)
```

```python
#geodesic-path-length
nx.shortest_path_length(G,source=0,target=4038)
```

```python
#Average shortest path length
nx.average_shortest_path_length(G)
```

```python
#clustering-coefficient
nx.clustering(G)
```

```python
#average-clustering-coefficient
nx.average_clustering(G)
```

```python
#strongly-connected-components
for c in nx.strongly_connected_components(G):
  print(c)
```

```python
#sparseness
nx.density(G)
```

```python
#k-connectedness
c = nx.k_components(G,flow_func=None)
print(c)
```
